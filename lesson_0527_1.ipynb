```python
import requests
url = 'https://tcgbusfs.blob.core.windows.net/dotapp/youbike/v2/youbike_immediate.json'
response = requests.request('GET',url)
if response.status_code == 200:
    print("連線成功")
    all_data = response.json()
    print(type(all_data))
else:
    print(f"連線失敗:{response.status_code}")
```

```python
# 表格模組pandas
import pandas as pd

dataFram = pd.DataFrame(data=all_data,columns=['sna','tot','sbi','sarea','mday','ar','bemp','act'])
dataFram.info( )

#選取列
dataFram[0:201]

```

```python
#選取欄位
dataFram[['sna', 'tot', 'sbi']]
```
